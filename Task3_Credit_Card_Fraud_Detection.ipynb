```python
# Check for missing values and handle them using imputation
imputer = SimpleImputer(strategy='median')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```

```python
# Use StandardScaler to scale the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_imputed)

# Convert scaled data back to DataFrame
df_scaled = pd.DataFrame(scaled_data, columns=df.columns)
```

```python
#addressing outlier using z-score
from scipy import stats
z_scores = np.abs(stats.zscore(df_scaled))
threshold = 3
outliers = (z_scores > threshold).all(axis=1)
df_no_outliers = df_scaled[~outliers]
```

```python
#Split the Dataset
X = df.drop('Class',axis=1)
y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.01,random_state=100)
```

```python
Model = RandomForestClassifier(random_state=100,n_jobs=-1,class_weight='balanced')
params = {'n_estimators':[100],
         'max_depth':[5,10,20,30,40],
         'min_samples_leaf':[10,20,50,100,200,400],
         'max_features':[0.1,0.2,0.3,0.4],
         'criterion':["gini"]}
df1 = GridSearchCV(estimator=Model,param_grid=params,verbose=1,n_jobs=-1,scoring='recall')
df1.fit(X_train,y_train)
```

```python
Model_best = df1.best_estimator_
```

```python
#evaluating model performance through accuracy, recall
from sklearn.metrics import accuracy_score, recall_score
y_train_pred = Model_best.predict(X_train)
print("Accuracy: ", accuracy_score(y_train, y_train_pred))
print("Recall: ", recall_score(y_train, y_train_pred))
```

```python
y_test_pred = Model_best.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, y_test_pred))
print("Recall: ", recall_score(y_test, y_test_pred))
```
